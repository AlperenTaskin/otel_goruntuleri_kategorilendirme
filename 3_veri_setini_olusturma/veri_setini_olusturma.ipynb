```python
import warnings
warnings.filterwarnings('ignore')
from keras.preprocessing.image import img_to_array, load_img
import glob
```

## Klasor Isimlerini Aliyoruz 

```python
klasorler = []
dosya_yolu = '../resimler/'
import os
for dirpath, dirnames, filenames in os.walk(dosya_yolu):
    print(dirpath)
    klasorler.append(dirpath.replace('../resimler/',''))
```

```python
klasorler.pop(0)
klasorler
```

```python
klasorler.remove('others')
```

```python
klasorler
```

## Klasorlerin IcÄ±ndeki Resimleri 224*224 Boyutunda Matrise Ceviriyoruz

```python
images = []
sayac = 0
# egitim verisinin girdi degerleri
def verileriOku(folders):
    for counter in range(len(folders)):
        for filename in glob.glob(dosya_yolu + folders[counter] + '/*.jpg'):
            img = load_img(filename, target_size=(224, 224))
            img = img_to_array(img)
            images.append(img)
            global sayac 
            sayac += 1
            print(str(sayac) + '. Resim ismi :' + filename)   
```

```python
verileriOku(klasorler)
```

## 224*224 Boyutunda ve 3 Kanalli Resimleri Verisetini Olusturmak Icin Duzlestiriyoruz

```python
import numpy as np
images = np.reshape(images, (4691,150528))
```

```python
import pandas as pd
df = pd.DataFrame(images)
df.columns = df.columns.astype(str)
```

```python
df.info()
```

```python
from fastparquet import write
write('resimler.parq', df)
```

```python
df.to_csv('resimler.csv')
```
