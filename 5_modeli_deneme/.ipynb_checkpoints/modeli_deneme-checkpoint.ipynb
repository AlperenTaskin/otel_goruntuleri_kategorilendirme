**Modeli Deneme**

```python
from keras.preprocessing.image import img_to_array, load_img
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
etiketler = ['yatak_odasi', 'restoran', 'banyo', 'veranda' , 'bina']
```

```python
def tahminEt(img):
  img = img_to_array(img)
  # keras modeline 4 boyutlu matris kabul ediyor
  img = np.reshape(img,(-1,224,224,3))
  img = img/255.0
  sonuc = model.predict(img)
  sonuc = etiketler[np.argmax(sonuc)]
  return sonuc
```

```python
def goruntule(img):
  plt.imshow(img)
  plt.axis('off')
```

# Modeli Deneme

```python
def goruntu_etiketlendir(resim_adi):
    resim = load_img('Test_Verileri/'+str(resim_adi)+'.jpg',target_size=(224,224))
   
    img = img_to_array(resim)
    img = np.reshape(img,(-1,224,224,3))
    img = img/255.0
    
    etiket = model.predict(img)
    etiket = etiketler[np.argmax(etiket)]
    
    fig=plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(1, 1, 1)
    
    ax.annotate('Kategori Tahmini : ' + etiket, (0,0), (5, -15), 
                xycoords='axes fraction', textcoords='offset points', va='top', size=15)
    plt.axis('off')
    plt.imshow(resim)
    
    plt.show()   
```

```python
for i in range(1,20):
    goruntu_etiketlendir(i)
```
